### Club Averages

```python



# --- separate the players from the keepers

players = bf[bf["Best Pos"] != "GK"]

players = players.drop( columns = ['Aer', 'Cmd',
       'Com', 'Ecc', 'Han', 'Kic', '1v1', 'Pun', 'Ref', 'TRO', 'Thr']) 

gk = bf[bf["Best Pos"] == "GK"]





# ---  Find the club averages for players

technical = ['Cor', 'Cro', 'Dri', 'Fin', 'Fir', 'Fre', 'Hea', 'Lon', 'L Th', 'Mar', 'Pas', 'Pen', 'Tck', 'Tec']
mental =  ['Agg', 'Ant', 'Bra', 'Cmp', 'Cnt', 'Dec', 'Det', 'Fla', 'Ldr', 'OtB', 'Pos', 'Tea', 'Vis', 'Wor']
physical = ['Acc', 'Agi', 'Bal', 'Jum', 'Nat.1', 'Pac', 'Sta', 'Str']
keeping = ['Aer', 'Cmd', 'Com', 'Ecc', 'Han', 'Kic', '1v1', 'Pun', 'Ref', 'TRO', 'Thr']


all_attrs = technical + mental + physical

player_club_averages = (
    players
    .groupby("Club")[all_attrs]
    .mean()
)

keeper_club_averages =  gk.groupby("Club")[keeping].mean()

player_club_averages[keeping] = keeper_club_averages[keeping]
player_club_averages["technical_avg"] = player_club_averages[technical].mean(axis=1)
player_club_averages["mental_avg"]    = player_club_averages[mental].mean(axis=1)
player_club_averages["physical_avg"]  = player_club_averages[physical].mean(axis=1)
player_club_averages["keeper_avg"]  = keeper_club_averages[keeping].mean(axis=1)

player_club_averages = player_club_averages.round(2)

player_club_averages



#club_average_keepers = gk.groupby(   ) 
```

### PCA

```python
import pandas as pd
```

```python



players = pd.read_csv("players.csv")

players = players[players["Best Pos"] != "GK"]

players = players.drop( columns = ['Aer', 'Cmd',
       'Com', 'Ecc', 'Han', 'Kic', '1v1', 'Pun', 'Ref', 'TRO', 'Thr']) 

# Remove non-numeric characters and convert to numeric
cols = ["Height", "Weight"]
players[cols] = players[cols].replace(r'[^\d.]', '', regex=True).apply(pd.to_numeric)

# Extract only the year from the string
players['Birth_Year'] = players['DoB'].str.extract(r'(\d{4})').astype(int)

# Subtract from the current year (2025)
players['DoB'] = 2025 - players['Birth_Year']

players.rename(columns={'DoB': 'Age'}, inplace=True)



players 
```

```python
# Keep only columns with numeric data types (int, float, etc.)
players = players.select_dtypes(include='number')

# Drop column: 'UID'
players = players.drop(columns=['UID' , 'Birth_Year'])

players 
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# -----------------------------
# 1. Standardize the data
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(players)

# -----------------------------
# 2. PCA keeping ~80% variance
# -----------------------------
pca = PCA(n_components=0.80)
X_pca = pca.fit_transform(X_scaled)

# -----------------------------
# 3. PCA scores as DataFrame
# -----------------------------
players_pca = pd.DataFrame(
    X_pca,
    index=players.index,
    columns=[f"PC{i+1}" for i in range(X_pca.shape[1])]
)

# -----------------------------
# 4. Loadings (feature contributions)
# -----------------------------
loadings = pd.DataFrame(
    pca.components_.T,
    index=players.columns,
    columns=players_pca.columns
)

# -----------------------------
# 5. Variance explained
# -----------------------------
explained_variance = pd.Series(
    pca.explained_variance_ratio_,
    index=players_pca.columns,
    name="Explained Variance"
)

cumulative_variance = explained_variance.cumsum()

# -----------------------------
# 6. Print summary
# -----------------------------
print(f"Number of components kept: {pca.n_components_}")
print("\nExplained variance by component:")
print(explained_variance)

print("\nCumulative explained variance:")
print(cumulative_variance)

```

### Training data creation

```python
train = pd.read_csv("results.csv")
```

```python
pc_cols = [f"PC{i}" for i in range(1, 14)]

club_pca = club_pca.reset_index()  # ensure Club is a column

for side in ["Home", "Away"]:
    train = (
        train
        .merge(club_pca, left_on=side, right_on="Club", how="left")
        .rename(columns={c: f"{c}_{side.lower()}" for c in pc_cols})
        .drop(columns="Club", errors="ignore")
    )
```

```python
train = train.rename(columns={
    "xG": "xG_home",
    "xG.1": "xG_away"
})

train
```

```python
# --- 2. Extract home / away goals from Score safely ---
# works for: "2–1", "2-1", "1–1 (ET)", etc.
goals = train["Score"].str.extract(r"(\d+)\D+(\d+)")

home_goals = pd.to_numeric(goals[0])
away_goals = pd.to_numeric(goals[1])

# --- 3. Encode result: 1 = home win, 0 = draw, -1 = home loss ---
train["result"] = (
    (home_goals > away_goals).astype(int)
    - (home_goals < away_goals).astype(int)
)


train
```

```python
train.to_csv("train.csv", index=False)
```
