```python
import pandas as pd
import os

# Load your CSV
df = pd.read_csv("attributes.csv")

# Folder containing photos (local)
photo_folder = "photos"

# Function to match each player to their photo filename
def get_photo_path(name):
    # You can adjust extensions if needed (.jpg / .jpeg)
    for ext in [".png", ".jpg", ".jpeg"]:
        filename = f"{name}{ext}"
        filepath = os.path.join(photo_folder, filename)
        if os.path.exists(filepath):
            return filepath
    return "photos/placeholder.png"


# Create the new column
df["Photo"] = df["Name"].apply(get_photo_path)

# Save the updated CSV
df.to_csv("attributes_with_photos.csv", index=False)

print("Done! Added Photo column.")
```

```python
import pandas as pd
import os

# Load your attributes file
df = pd.read_csv("attributes.csv")

# Folder where all images are stored
photo_folder = "photos"

def get_image_path(base_name):
    """
    Given a base name (player or club), return a valid image path.
    Tries .png, .jpg, .jpeg. Returns '' if nothing found.
    """
    if pd.isna(base_name):
        return ""

    for ext in [".png", ".jpg", ".jpeg"]:
        filename = f"{base_name}{ext}"
        filepath = os.path.join(photo_folder, filename)
        if os.path.exists(filepath):
            return filepath

    return "photos/placeholder.png"   # Safe fallback

# Create the Photo column (matches Name column)
df["Photo"] = df["Name"].apply(get_image_path)

# Create the ClubBadge column (matches Club column)
df["ClubBadge"] = df["Club"].apply(get_image_path)

# Save result
df.to_csv("attributes_with_images.csv", index=False)
```
